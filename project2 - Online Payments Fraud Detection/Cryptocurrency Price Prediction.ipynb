```python
import yfinance as yf
```

```python
from datetime import date,timedelta
today=date.today()
```

```python
d1=today.strftime('%Y-%m-%d')
end_date=d1
print(end_date)
```

```python
d2=date.today()-timedelta(days=2*365)
start_date=d2
print(d2)
```

```python
data=yf.download('BTC-USD',
                 start=start_date,
                 end=end_date,
                 progress=False)
```

```python
data
```

```python
data['Date']=data.index
```

```python
data=data[['Date','Open','High','Low','Close','Adj Close','Volume']]
```

```python
data
```

```python
data.reset_index(drop=True,inplace=True)
data.head()
```

```python
data.shape
```

```python
x=data['Date']
open=data['Open']['BTC-USD']
high=data['High']['BTC-USD']
low=data['Low']['BTC-USD']
close=data['Close']['BTC-USD']
```

```python
import plotly.graph_objects as go 
figure=go.Figure(data=[go.Candlestick(x=x,
                                      open=open,
                                      high=high,
                                      low=low,
                                      close=close)])

figure.show()
```
